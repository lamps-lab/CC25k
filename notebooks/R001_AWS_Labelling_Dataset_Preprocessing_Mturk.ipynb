```python
import nltk
import re
import numpy as np

def identify_citation_marks_round_1(text):

    counts_from_each = {'numeric_single_mark':0, 
                        'numeric_multiple_mark':0, 
                        'numeric_multiple_mark_adjacent':0, 
                        'text_single_mark_1_auth':0, 
                        'text_single_mark_2_auth':0,
                        'text_multiple_type_2':0} 

    citation_pattern_numeric_single_mark = r'\[\d+\]'  #-- [17]
    citation_pattern_numeric_multiple_mark = r'\[\d+(?:,\s*\d+)*\]'  #-- [17]
    citation_pattern_numeric_multiple_mark_adjacent = r'(\[\d+\]){2,}'  #-- [17][18][19]

    citation_pattern_text_single_mark_1_auth = r'\([\w*\s*]*\.*,*\s*\d{4}\)'
    citation_pattern_text_single_mark_2_auth = r'\([\w*\s*]*&[\w*\s*]*,*\s*\d{4}\)'
    
    citation_pattern_text_multiple_type_2 = r'\(([\w*\s*]*&[\w*\s*]*,*\s*\d{4}[a-z]?;*\s*)*([\w*,*\s*]*\.*,*\s*\d{4}[a-z]?;\s*)+([\w*\s*]*&[\w*\s*]*,*\s*\d{4}[a-z]?;*\s*)*([\w*,*\s*]*\.*,*\s*\d{4}[a-z]?)*\)'
        
    numeric_marks_1 = re.findall(citation_pattern_numeric_single_mark,text)
    numeric_marks_2 = re.findall(citation_pattern_numeric_multiple_mark,text)
    numeric_marks_2 = set(numeric_marks_2).difference(set(numeric_marks_1))
    numeric_marks_3 = re.findall(citation_pattern_numeric_multiple_mark_adjacent,text)
    
    text_marks_1 = re.findall(citation_pattern_text_single_mark_1_auth,text) 
    text_marks_2 = re.findall(citation_pattern_text_single_mark_2_auth,text) 
    text_marks_4 = re.findall(citation_pattern_text_multiple_type_2,text)
    
    counts_from_each['numeric_single_mark'] = len(numeric_marks_1)
    counts_from_each['numeric_multiple_mark'] = len(numeric_marks_2)
    counts_from_each['numeric_multiple_mark_adjacent'] = len(numeric_marks_3)
    
    counts_from_each['text_single_mark_1_auth'] = len(text_marks_1)
    counts_from_each['text_single_mark_2_auth'] = len(text_marks_2)
    counts_from_each['text_multiple_type_2'] = len(text_marks_4)
                        
                        
    numeric_marks_1.extend(numeric_marks_2)
    numeric_marks_1.extend(numeric_marks_3)
    numeric_marks_1.extend(text_marks_1)
    numeric_marks_1.extend(text_marks_2)
    numeric_marks_1.extend(text_marks_4)

    return set(numeric_marks_1), counts_from_each
```

```python
def identify_citation_marks_round_2(text, author):
    
    author_names = author.split()

    if author_names[-1] in text:
        return True
    
    return False
```

```python
def check_words_in_sentence(sentence, word_list):
    # Tokenize the sentence
    tokens = nltk.word_tokenize(sentence)
    # Create a set of tokens for faster lookup
    token_set = set(tokens)
    # Check if each word in the list exists in the token set
    for word in word_list:
        if word in token_set:
            return True
    return False

# List of words to check
word_list = ["but", "instead", "however","yet","nevertheless","nonetheless","still", "though", "although", 
             "despite", "notwithstanding", "even so", "in spite of", "on the other hand", "in contrast", 
             "conversely", "whereas", "whilst", "alternatively", "on the contrary", "rather", 
             "alternative","regardless"]
```

```python
def is_correct_citation_mark_found(citation_list, author):
    for citation_mark in citation_list:
        if author in citation_mark: 
            return (True, citation_mark)
    return (False, 'test')
```

```python
citation_mark_counts_per_context = {'_0_cit_marks':0,
                                    '_1_cit_marks':0,
                                    '_2_cit_marks':0,
                                    '_3_cit_marks':0,
                                    '_4_cit_marks':0,
                                    '_5_cit_marks':0,
                                    '_6plus_cit_marks':0}

contradictory_words_in_context = {'words_exist':0,
                                    'no_words':0}

numeric_citation_style = 0
text_based_style = 0 
both_styles_in_same_context_count = 0
adjacent_count = 0 # [1][2][3]
numeric_count_multi = 0 #[1][2][3,4,5]
text_count_multi = 0 # (Ya et al., 2022; Dai & Hang, 2021; Zhu et al., 2022) Samples RestainNet (Zhao et al., 2022)
unidentifed_citation_count = 0
contexts_with_no_citation_mark = 0
citation_marks_found_from_round_2 = 0

csv_file_list = glob.glob("AWS_labelling/Mechanical_Turk/dataset_pre_processing/file_key_first_author/*.csv")

total_context = 0
total_files = 0
for file in csv_file_list:
    df_new = pd.DataFrame(columns=['index','context','file_key','first_author'])
    key_ = os.path.basename(file)[:-4]
    file_name = [i for i in file_list_json if key_ in i][0]
    author_name = ''

    with open(file_name) as f:
        cite_count = 0
        data = json.load(f)
        author_name = data['authors'][0]['name'].split(' ')[-1]

    try:
        df = pd.read_csv(file)
    except:
        print('File is EMPTY')
        
        with open(r'AWS_labelling/Mechanical_Turk/dataset_pre_processing/citation_marks_test.txt','a') as f:
            f.write(f'\n\n\n{file}: File is EMPTY\n\n')
    else:
        with open(r'AWS_labelling/Mechanical_Turk/dataset_pre_processing/citation_marks_test.txt','a') as f:
            f.write(f'\n\n\n{file}:\n')
        
        total_files += 1
        print(len(df))
        total_context += len(df)
        
        for i in range(len(df['index'])):
            cit_marks, counts = identify_citation_marks_round_1(df['context'][i])
            count = counts['numeric_single_mark'] + counts['numeric_multiple_mark'] \
            + counts['text_single_mark_1_auth'] + counts['text_single_mark_2_auth'] + counts['text_multiple_type_2']
                        
            numeric_only_count = counts['numeric_single_mark'] + counts['numeric_multiple_mark']
            text_only_count = counts['text_single_mark_1_auth'] + counts['text_single_mark_2_auth']+ counts['text_multiple_type_2']
            
            # citation styles
            if (numeric_only_count > 0) and (text_only_count == 0):
                numeric_citation_style += 1                
                text = df['context'][i]
                
            elif (numeric_only_count == 0) and (text_only_count > 0):
                text_based_style += 1                
                text = df['context'][i]                
                ret_, mark = is_correct_citation_mark_found(cit_marks, author_name)
                
                if ret_ == True:
                    df.at[i, 'first_author'] =  'Cited Paper: ' + mark
                
            elif (numeric_only_count > 0) and (text_only_count > 0):
                both_styles_in_same_context_count += 1
            
            if count == 1:
                citation_mark_counts_per_context['_1_cit_marks'] += 1
                df_new.loc[len(df_new.index)] = list(df.loc[i]) #[df[0][i]]                 
            elif count == 2:
                citation_mark_counts_per_context['_2_cit_marks'] += 1
            elif count == 3:
                citation_mark_counts_per_context['_3_cit_marks'] += 1
            elif count == 4:
                citation_mark_counts_per_context['_4_cit_marks'] += 1
            elif count == 5:
                citation_mark_counts_per_context['_5_cit_marks'] += 1
            elif count == 0:
                citation_mark_counts_per_context['_0_cit_marks'] += 1

                if 'et al.' in str(df['context'][i]):
                    unidentifed_citation_count +=1
                    df_new.loc[len(df_new.index)] = list(df.loc[i])
                    
                    with open(r'AWS_labelling/Mechanical_Turk/dataset_pre_processing/citation_marks_test.txt','a',encoding="utf-8") as f:
                        f.write(df['context'][i])
                        f.write('\n\n')
                else:                        
                    ret_r2 = identify_citation_marks_round_2(str(df['context'][i]), author_name)
                    ret_, mark = is_correct_citation_mark_found(cit_marks, author_name)
                    
                    if ret_r2 == True:
                        if ret_ == True:
                            df.at[i, 'first_author'] =  'Cited Paper: ' + mark
                    if ret_r2 == True and author_name != 'Kaiming He':
                        citation_marks_found_from_round_2 +=1
                        df_new.loc[len(df_new.index)] = list(df.loc[i])
                    else:
                        contexts_with_no_citation_mark +=1
            else:
                citation_mark_counts_per_context['_6plus_cit_marks'] += 1
                           
            # scenarios for multiple citations in single context 
            if counts['numeric_single_mark'] + counts['numeric_multiple_mark'] >1:  # [1,3,4] sfdhbf [5]
                numeric_count_multi += 1
                        
            if counts['text_single_mark_1_auth'] + counts['text_single_mark_2_auth'] + counts['text_multiple_type_2'] > 1: # (Ya et al., 2022; Dai & Hang, 2021; Zhu et al., 2022) Samples RestainNet (Zhao et al., 2022)
                text_count_multi += 1
                df_new.loc[len(df_new.index)] = list(df.loc[i])                
                
            if counts['numeric_multiple_mark_adjacent'] != 0: #[1][3][4]
                adjacent_count += 1

            if count > 1:
                result = check_words_in_sentence(df['context'][i], word_list)
                if result is True:
                    contradictory_words_in_context['words_exist'] += 1
                else:
                    contradictory_words_in_context['no_words'] += 1 

    path_ = os.path.join('AWS_labelling','Mechanical_Turk','dataset_pre_processing','file_key_first_author_filtered',os.path.basename(file))    
    df_new.to_csv(path_,index=False)
    print(f'new count: {len(df_new)}\n')

print(f'contexts with 1 citation marks: {citation_mark_counts_per_context["_1_cit_marks"]}')
print(f'contexts with 2 citation marks: {citation_mark_counts_per_context["_2_cit_marks"]}')
print(f'contexts with 3 citation marks: {citation_mark_counts_per_context["_3_cit_marks"]}')
print(f'contexts with 4 citation marks: {citation_mark_counts_per_context["_4_cit_marks"]}')
print(f'contexts with 5 citation marks: {citation_mark_counts_per_context["_5_cit_marks"]}')
print(f'contexts with 5+ citation marks: {citation_mark_counts_per_context["_6plus_cit_marks"]}')
print(f'contexts with unidentified citation marks: {citation_mark_counts_per_context["_0_cit_marks"]}')
print(f'contexts with unidentified citation marks [contains et al. but not identifed with current regex]: {unidentifed_citation_count}')
print(f'contexts with unidentified citation marks [no citation marks]: {contexts_with_no_citation_mark}')
print(f'All counts: {sum(citation_mark_counts_per_context.values())}')

print('total_context : ',total_context)
print('total_files : ',total_files)

print('\n\n\nAdditional Details:\n')
print('Citation Styles:')
print(f'Contexts having only IEEE like style: {numeric_citation_style}')
print(f'Contexts having only APA like style: {text_based_style}')
print(f'Contexts having both IEEE and APA like style: {both_styles_in_same_context_count}')

print('Scenarios for multiple citations in a single context')
print(f'Ex: [1,3,4] text text text [5]: {numeric_count_multi}')
print(f'Ex: [1][3][4]: {adjacent_count}')
print(f'Ex: # (Ya et al., 2022; Dai & Hang, 2021) Samples RestainNet (Zhao et al., 2022) : {text_count_multi}') 

print("\n\n")
print('contexts with contradictory words when there are multiple citations in a single context')
print(f'contexts with contradictory words: {contradictory_words_in_context["words_exist"]}')
print(f'contexts without contradictory words: {contradictory_words_in_context["no_words"]}')
print(f'All counts: {sum(contradictory_words_in_context.values())}')
```
