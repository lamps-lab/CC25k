# word cloud

```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

file_path = r"dataset/cc25k_dataset_m1_m5.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)

# Initialize the figure and subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

# Categories for which to generate word clouds
categories = ["neutral", "positive", "negative"]

# Define a custom set of words to exclude (you can expand this list as needed)
custom_stopwords = set(STOPWORDS).union({"et al", "et", "al", "and", "a"})  # Adding some extra words

# Generate word clouds per category and plot them in subplots
for ax, category in zip(axes, categories):
    category_texts = " ".join(df[df["majority_vote"] == category]["input_context"])
    wordcloud = WordCloud(width=800, height=400, background_color="white", stopwords=custom_stopwords).generate(category_texts)
    
    # Display the wordcloud
    ax.imshow(wordcloud, interpolation="bilinear")
    ax.axis("off")
    ax.set_title(f"Word Cloud for {category.capitalize()} Category")

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 1])
plt.savefig(r'plots/word_cloud.pdf')
plt.show()
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from nltk.tokenize import word_tokenize, sent_tokenize

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk

# Load your dataset (replace with your actual file path)
file_path = r"dataset/cc25k_dataset_m1_m5.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)

# Function to calculate word count using nltk.word_tokenize
def word_count(text):
    words = word_tokenize(text)  # Tokenize text into words
    return len(words)

# Function to calculate sentence count using nltk.sent_tokenize
def sentence_count(text):
    sentences = sent_tokenize(text)  # Tokenize text into sentences
    return len(sentences)

# Add word count and sentence count for each context
df['word_count'] = df['input_context'].apply(word_count)
df['sentence_count'] = df['input_context'].apply(sentence_count)

# Create bins for word count distribution
word_bins = range(0, 81, 4)  # Bins from 0 to 80 with a step of 4

# Create a figure for the main plot
fig, ax = plt.subplots(figsize=(8, 6))

# Categories for visualization
categories = ["neutral", "positive", "negative"]
colors = ['#86bbd8', '#6bbf59', '#a4243b']  # Colors for each category skyblue salmon a4133c b5e2fa lightgreen 69b578 a9e5bb
 
# Plot the distribution of word counts per context for each category with fixed bin positions
for category, color in zip(categories, colors):
    category_df = df[df["majority_vote"] == category]
    sns.histplot(category_df['word_count'], bins=word_bins, color=color, label=category, kde=False, kde_kws={'bw_adjust': 0.5},  ax=ax, alpha=0.8, edgecolor='#343a40')

# Set the x-axis limit to stop at word count 80
ax.set_xlim(0, 80)

# Add titles, labels, and legend
ax.set_xlabel("Word Count per Context")
ax.set_ylabel("Frequency (Log Scale)")
ax.legend(title="Categories", loc='upper left')

# Create inset axes for sentence count distribution
axins = inset_axes(ax, width="30%", height="30%", loc='upper right',
                   bbox_to_anchor=(-0.05, -0.1, 1.0, 1.0), bbox_transform=ax.transAxes)

# Plot the sentence count distribution as a bar plot in the inset
for category, color in zip(categories, colors):
    category_df = df[df["majority_vote"] == category]
    sns.histplot(category_df['sentence_count'], bins=range(0, 9), color=color, label=category, discrete=True, ax=axins, alpha=1.0, edgecolor='#343a40')  # Set alpha=1 to remove transparency

# Set the x-axis limit for the inset plot to stop at sentence count 8
axins.set_xlim(0, 9)
axins.set_xticks(range(0, 10))  # Ensure only integer ticks from 0 to 8 are shown

# Set the y-axis of the inset plot to log scale
axins.set_yscale('log')

# Add titles and labels to the inset plot
axins.set_title("Sentence Count Distribution")
axins.set_xlabel("Sentence Count")
axins.set_ylabel("Frequency (Log Scale)")

# Show the plot
plt.tight_layout()
plt.savefig(r'plots/word_count_distribution_with_inset_log.pdf')  # Save the plot as a PDF (optional)
plt.show()

# Calculate average word count per context
avg_word_count = df['word_count'].mean()

# Calculate average sentence count per context
avg_sentence_count = df['sentence_count'].mean()

# Display the results
print(f"Average word count per context: {avg_word_count:.2f}")
print(f"Average sentence count per context: {avg_sentence_count:.2f}")
```
